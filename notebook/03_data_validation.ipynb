```python
import pandas as pd

# ─────────────────────────────────────────
# Load Data
# ─────────────────────────────────────────
df = pd.read_csv('../data///customer_sentiment_list.csv')

print("=" * 60)
print("DATA VALIDATION REPORT")
print("=" * 60)

# Check 1 — Total Reviews
print(f"\n✅ Total Reviews: {len(df):,}")

# Check 2 — Sentiment Distribution
print("\n✅ Sentiment Distribution:")
sentiment_dist = df['vader_sentiment'].value_counts()
sentiment_pct  = df['vader_sentiment'].value_counts(normalize=True) * 100
for sentiment in sentiment_dist.index:
    print(f"   {sentiment}: {sentiment_dist[sentiment]:,} ({sentiment_pct[sentiment]:.2f}%)")

# Check 3 — Average Sentiment Score
print(f"\n✅ Avg Sentiment Score: {df['vader_score'].mean():.4f}")

# Check 4 — Sentiment Alignment
print(f"\n✅ Sentiment Alignment %: {(df['sentiment_alignment'] == 'Match').mean() * 100:.2f}%")

# Check 5 — Top Complaint Theme
print(f"\n✅ Theme Distribution:")
theme_dist = df['theme'].value_counts()
for theme, count in theme_dist.items():
    print(f"   {theme}: {count:,}")

# Check 6 — Negative Review Rate
negative = df['vader_sentiment'].str.contains('Negative', na=False).sum()
total    = len(df)
print(f"\n✅ Negative Review Rate: {negative/total*100:.2f}%")

# Check 7 — Missing values
print(f"\n✅ Missing Values:")
print(df.isnull().sum()[df.isnull().sum() > 0])

print("\n" + "=" * 60)
print("VALIDATION COMPLETE")
print("=" * 60)
```
